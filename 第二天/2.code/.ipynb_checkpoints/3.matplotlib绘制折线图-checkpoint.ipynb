#  绘制基本折线图

```python
# 0.准备数据
x = range(60)
y = [random.uniform(15, 20) for i in x]

# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.plot(x, y)

# 2.1 添加x轴,y轴刻度
x_ticks_label = ["11点{}分".format(i) for i in range(60)]
# print(x_ticks_label)
y_ticks = range(40)

# plt.xticks(x_ticks_label[::5])  # 报错,直接显示字符串不可以
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 2.2 添加网格显示
plt.grid(True, linestyle="--", alpha=0.7)

# 2.3 添加x,y轴坐标描述和添加标题
plt.title("温度变化图", fontsize=20)
plt.xlabel("时间")
plt.ylabel("温度")

# 3.图像显示
plt.show()
```

# 绘制多个折现

```python
# 0.准备数据
x = range(60)
y_shanghai = [random.uniform(15, 20) for i in x]
y_beijing = [random.uniform(5,15) for i in x]

# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.plot(x, y_shanghai, label="上海", color="r")
plt.plot(x, y_beijing, linestyle = "--", label="北京")

# 2.1 添加x轴,y轴刻度
x_ticks_label = ["11点{}分".format(i) for i in range(60)]
# print(x_ticks_label)
y_ticks = range(40)

# plt.xticks(x_ticks_label[::5])  # 报错,直接显示字符串不可以
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 2.2 添加网格显示
# plt.grid(True, linestyle="--", alpha=0.7)
plt.grid(False)


# 2.3 添加x,y轴坐标描述和添加标题
plt.title("温度变化图", fontsize=20)
plt.xlabel("时间")
plt.ylabel("温度")

# 2.4 显示图例
plt.legend(loc ="best")

# 3.图像显示
plt.show()
```

```python
# 0.准备数据
x = range(60)
y_shanghai = [random.uniform(15, 20) for i in x]
y_beijing = [random.uniform(5,15) for i in x]

# 1.创建画布
# plt.figure(figsize=(20, 8), dpi=100)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)

# 2.绘制图像
# plt.plot(x, y_shanghai, label="上海", color="r")
# plt.plot(x, y_beijing, linestyle = "--", label="北京")
axes[0].plot(x, y_shanghai, label="上海", color="r")
axes[1].plot(x, y_beijing, linestyle = "--", label="北京")


# # 2.1 添加x轴,y轴刻度
x_ticks_label = ["11点{}分".format(i) for i in range(60)]
# print(x_ticks_label)
y_ticks = range(40)

# # plt.xticks(x_ticks_label[::5])  # 报错,直接显示字符串不可以
# plt.xticks(x[::5], x_ticks_label[::5])
# plt.yticks(y_ticks[::5])
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])

# # 2.2 添加网格显示
# # plt.grid(True, linestyle="--", alpha=0.7)
# plt.grid(False)
axes[0].grid(True, linestyle="--", alpha=0.7)
axes[1].grid(True, linestyle="--", alpha=0.7)

# # 2.3 添加x,y轴坐标描述和添加标题
# plt.title("温度变化图", fontsize=20)
# plt.xlabel("时间")
# plt.ylabel("温度")
axes[0].set_title("温度变化图", fontsize=20)
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[1].set_title("温度变化图", fontsize=20)
axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")


# # 2.4 显示图例
# plt.legend(loc ="best")
axes[0].legend()
axes[1].legend()

# 3.图像显示
plt.show()
```
