```python
#生成均匀分布值得验证

# 0.生成数据
x = np.random.uniform(0, 1, 10000000)

# 1.画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制
plt.hist(x, bins=1000)

# 3.展示
plt.show()
```

```python
np.random.normal(1.75, 0.1, (4, 10))
```

```python
# 创建正态分布图像
# 0.数据
x = np.random.normal(0, 1, 100000000)

# 1.画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制
plt.hist(x, bins=1000)

# 3.显示
plt.show()
```
